```python
import pandas as pd
import matplotlib.pyplot as plt 

dados_btc = pd.read_csv(
                    "btc.csv",
                    sep=',',
                    index_col='Date',
                    usecols=['Date','High','Low','Close'])
```

# Aqui no primeiro gráfico de linhas é possível observar o aumento no preço do bitcoin entre 2013-2021: 

```python
plt.plot(dados_btc)
```

# Já no segundo gráfico, ainda sobre o bitcoin, é possível analisar melhor a variação apenas no período das eleições americanas em 2020 e começo de 2021: 

```python
dados_btc_fix = dados_btc[dados_btc["Low"] <= 18800.000000].index
dados_novo = dados_btc.drop(dados_btc_fix) 
plt.plot(dados_novo)
```
