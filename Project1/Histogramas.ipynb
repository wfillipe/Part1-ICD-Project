```python
import pandas as pd
import matplotlib.pyplot as plt 

dados_btc = pd.read_csv(
                    "btc.csv",
                    sep=',',
                    index_col='Date',
                    usecols=['Date','High','Low','Close'])
    
dados_btc
```

```python
dados_btc_fix = dados_btc[dados_btc["Low"] <= 18800.000000].index
dados_novo = dados_btc.drop(dados_btc_fix) 
dados_novo
```

# O primeiro histograma é possível analisar a alta variação de preço no bitcoin (em dólares) no período de 2020: 

```python
dados_novo.hist(column='Close', bins=100)
```

```python
dados_USD = pd.read_csv(
                        "USD.csv",
                        sep=',',
                        index_col='Data',
                        usecols=['Data', 'Máxima','Mínima','Var%'])
dados_USD
```

```python
# Não estava conseguindo rodar o histograma do data frame "dados_USD" porque as colunas não estavam categorizadas como números,
# nesta célula o problema é resolvido

df = pd.DataFrame(dados_USD)
df['Máxima'] = df['Máxima'].astype(float)
dados_USDnovo = df 
dados_USDnovo
```

# Aqui no segundo histograma está destacado o preço máximo do dólar (em reais) no período de 2020 (ano de eleições americanas):

```python
dados_USDnovo.hist(column='Máxima', bins=100)
```
